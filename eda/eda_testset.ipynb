```python
import os
```

```python
dataset1_dir = '../data/raw/CUSTOM DATASET/BACK VIEW'
dataset2_dir = '../data/raw/CUSTOM DATASET/SIDE VIEW'

def count_files_in_subfolders(root):
    counts = {}
    total = 0
    for dirpath, dirnames, filenames in os.walk(root):
        rel = os.path.relpath(dirpath, root)
        if rel == '.':
            rel = os.path.basename(root) or root

        file_count = len([
            f for f in filenames 
            if os.path.isfile(os.path.join(dirpath, f))
        ])

        if file_count > 0:       
            counts[rel] = file_count
            total += file_count

    return counts, total

for name, path in [('dataset1', dataset1_dir), ('dataset2', dataset2_dir)]:
    if not os.path.exists(path):
        print(f"{name}: đường dẫn không tồn tại: {path}")
        continue

    counts, total = count_files_in_subfolders(path)
    print(f"\n{name} ({path}) - total files: {total}")
    for sub, c in sorted(counts.items()):
        print(f"  {sub}: {c}")
```
